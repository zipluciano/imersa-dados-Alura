# Acompanhamento da aula

```python
# Importação da biblioteca Pandas
import pandas as pd

# Atribuindo à variável "url", o link da base de dados
url = "https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true"

# Importando a base de dados do GitHub, usando o parâmetro "compression", para decompactar o arquivo
dados = pd.read_csv(url, compression = "zip")

# Visualização da base de dados
dados
```

```python
# Função "unique" serve para descobrir quantas categorias ou itens distintos existem em uma coluna de um DataFrame
# Retorna um array com os dados que não se repetem
dados['tratamento'].unique()
```

```python
# O "value_counts()" está relacionado ao "unique()", pois o "unique()" descreve quais dados não repetidos existem em uma coluna
# Já o "value_counts()" conta quantas vezes esses valores aparecem
dados['tratamento'].value_counts()
```

```python
# Para normalizar os dados de uma coluna, a fim de encontrar uma proporção, pode-se utilizar o parâmetro "nomarlize" dentro do "value_counts()"
dados['tratamento'].value_counts(normalize = True)
```

```python
# A função plot permite plotar um gráfico, dependendo do tipo "pizza(pie)", "barras(bar)", etc
dados['tempo'].value_counts().plot.bar()
```

```python
# Selicionar dados com base em uma condição
dados['g-0'] > 0
# Retorna uma Series, com True ou False, para a condição

# Selcionando somente as linhas do DataFrame que corresponde à condição anterior
dados_filtro = dados[dados['g-0'] > 0]
```

## Desafio 05 ✔️

```python
# Existe uma função chamada "Query", que retorna as linhas de um DataFrame com base uma condição
# Página da função na documentação: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html

# Realizando a mesma query feita em aula, porém através da função Query
# Consulta: dados[dados['g-0'] > 0]
# Como o nome da coluna possui um hífen, é necessário inserir crase no nome da coluna, dentro das aspas (simples ou duplas)
dados_filtro = dados.query('`g-0` > 0')
```

## Desafio 06 ✔️

```python
# Removendo os hífens dos nomes de colunas, com a função "DataFrame.columns.str.replace"
nova_coluna = list(dados.columns.str.replace("-", ""))

# Substituindo os nomes de colunas com hífen pela nova lista de colunas
dados.columns = nova_coluna
dados
```
