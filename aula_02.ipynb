```python
# Criando uma cópia do DataFrame, para fazer modifacações, matendo o original na memória
dados2 = dados.copy()
```

```python
# Dicionário contendo a mudança de nomes da coluna "droga", fazendo de "droga" para "composto"
mapa = {'droga': 'composto'}

# Substituindo o nome da coluna através da função "rename". Passando o valor "True" para o parâmetro "inplace", dessa forma
# modifica-se o DataFrame permanentemente
dados2.rename(columns = mapa, inplace = True)
```

```python
# Importando a biblioteca "Seaborn", para plotagem de gráficos
import seaborn as sns
```

```python
# Plotando os 5 tipos de compostos mais usados, para isso é necessário antes fazer uma manipulação dos dados
# Selecionando os 5 tipos de compostos mais usados
dados2['composto'].value_counts()[0:5]

# Tomando somente o nome dos 5 compostos e agora ignorando a ocorrência
cod_compostos = dados2['composto'].value_counts()[0:5].index

# Selecionando no DataFrame as linhas que possuem um dos 5 compostos, através da função "query"
# Para fazer query é necessário adicionar na frente da condição o "@", pois a condição é um objeto externo ao DataFrame
compost_5 = dados2.query("composto in @cod_compostos")
```

```python
# Plotando o gráfico

# Importando o Matplotlib para aumentar o tamanho do gráfico
import matplotlib.pyplot as plt

# Aumentando o tamanho do gráfico através do Matplotlib
plt.figure(figsize = (10, 8))

# Ajustando o "countplot" para salvar as configurações de imagens do Seaborn padrão
sns.set()

## No Seaborn basta passar o eixo desejado como valor do parâmetro "x" e informar a fonte dos dados
sns.countplot(x = 'composto', data = compost_5)
```

```python
# Modificando o título do gráfico. Antes de modificar, uma boa prática é armazenar o gráfico em uma variável,
# geralmente chamada de "ax". Feito através do método "set_title"
plt.figure(figsize = (8, 6))
ax = sns.countplot(x = 'composto', data = compost_5)
ax.set_title("Top 5 compostos")
# "set_*" traz modificações possíveis de se fazer no gráfico

# Apresentar o gráfico
plt.show()
```

```python
# Para variáveis contínuas, um tipo de gráfico indicado é o histograma, pode-se chamar ele pelo método "hist"
dados2['g0'].hist()
```

```python
# Como o parâmetro "bins", é possível definir a largura das barras, tornando o gráfico próximo ao de uma curva
dados2['g0'].hist(bins = 100)
# O gráfico do "g0", com bins setado para 100 tem uma certa semelhança com uma distribuição normal de probabilidades.
# Testes são necessários para confirmar se realmente segue a distribuição
```

```python
# Já o gráfico do "c9", com bins setado para 100 tem uma certa semelhança com uma distribuição Gumbel de probabilidades.
# Testes são necessários para confirmar se realmente segue a distribuição
dados2['c9'].hist(bins = 100)
```

```python
# Algumas funções de estatística descritiva que existem no Pandas
dados2['g0'].min() # Retorna o menor valor da coluna
dados2['g0'].max() # Retorna o maior valor da coluna
dados2['g0'].mean() # Retorna a média da coluna
dados2['g0'].count() # Retorna quantos registros existem na coluna

# Método "describe" retorna um DataFrame com as estatísticas descritivas de todo o DataFrame consultado ou de um slice dele
dados2.describe()
```

### Desafio 06 ✔️

```python
# Análise de dois tipos de genes e dois tipos de linhagem de células aleatórios, utilizando histograma e Boxplot
sns.set()
fig, axes = plt.subplots(nrows = 4, ncols = 2, figsize = (20, 18))
fig.suptitle("Análise de genes e linhagem de células", fontsize = 30, color = "Blue")

hist_g1 = sns.histplot(ax = axes[0,0], data = dados2, x = 'g75', bins = 100, color = "Red")
hist_g2 = sns.histplot(ax = axes[1,0], data = dados2, x = 'g500', bins = 100, color = "Green")
box_g1 = sns.boxplot(ax = axes[0, 1], data = dados2, x = 'g75', color = "Red")
box_g2 = sns.boxplot(ax = axes[1, 1], data = dados2, x = 'g500', color = "Green")
hist_c1 = sns.histplot(ax = axes[2,0], data = dados2, x = 'c8', bins = 100, color = "Blue")
hist_c2 = sns.histplot(ax = axes[3,0], data = dados2, x = 'c42', bins = 100, color = "Orange")
box_c1 = sns.boxplot(ax = axes[2, 1], data = dados2, x = 'c8', color = "Blue")
box_c2 = sns.boxplot(ax = axes[3, 1], data = dados2, x = 'c42', color = "Orange")

hist_g1.set_xlabel(""), hist_g1.set_ylabel("Gráficos g75", fontsize = 18, color = "Red"), box_g1.set_xlabel(""), box_g1.set_ylabel("")
hist_g2.set_xlabel(""), hist_g2.set_ylabel("Gráficos g500", fontsize = 18, color = "Red"), box_g2.set_xlabel(""), box_g2.set_ylabel("")
hist_c1.set_xlabel(""), hist_c1.set_ylabel("Gráficos c8", fontsize = 18, color = "Red"), box_c1.set_xlabel(""), box_c1.set_ylabel("")
hist_c2.set_xlabel(""), hist_c2.set_ylabel("Gráficos c42", fontsize = 18, color = "Red"), box_c2.set_xlabel(""), box_c2.set_ylabel("")
plt.show()
# É possível observar que a cauda mais longa do histograma aparece como outlier no Boxplot
```
