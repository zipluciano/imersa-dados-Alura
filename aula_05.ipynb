```python
# Criando uma cópia do DataFrame, para fazer modifacações, matendo o original na memória
dados2 = dados.copy()
```

```python
# Dicionário contendo a mudança de nomes da coluna "droga", fazendo de "droga" para "composto"
mapa = {'droga': 'composto'}

# Substituindo o nome da coluna através da função "rename". Passando o valor "True" para o parâmetro "inplace", dessa forma
# modifica-se o DataFrame permanentemente
dados2.rename(columns = mapa, inplace = True)
```

```python
# Importando a biblioteca "Seaborn", para plotagem de gráficos
import seaborn as sns
```

```python
# Plotando os 5 tipos de compostos mais usados, para isso é necessário antes fazer uma manipulação dos dados
# Selecionando os 5 tipos de compostos mais usados
dados2['composto'].value_counts()[0:5]

# Tomando somente o nome dos 5 compostos e agora ignorando a ocorrência
cod_compostos = dados2['composto'].value_counts()[0:5].index

# Selecionando no DataFrame as linhas que possuem um dos 5 compostos, através da função "query"
# Para fazer query é necessário adicionar na frente da condição o "@", pois a condição é um objeto externo ao DataFrame
compost_5 = dados2.query("composto in @cod_compostos")
```

```python
# Plotando o gráfico

# Importando o Matplotlib para aumentar o tamanho do gráfico
import matplotlib.pyplot as plt

# Aumentando o tamanho do gráfico através do Matplotlib
plt.figure(figsize = (10, 8))

# Ajustando o "countplot" para salvar as configurações de imagens do Seaborn padrão
sns.set()

## No Seaborn basta passar o eixo desejado como valor do parâmetro "x" e informar a fonte dos dados
sns.countplot(x = 'composto', data = compost_5)
```

```python
# Modificando o título do gráfico. Antes de modificar, uma boa prática é armazenar o gráfico em uma variável,
# geralmente chamada de "ax". Feito através do método "set_title"
plt.figure(figsize = (8, 6))
ax = sns.countplot(x = 'composto', data = compost_5)
ax.set_title("Top 5 compostos")
# "set_*" traz modificações possíveis de se fazer no gráfico

# Apresentar o gráfico
plt.show()
```

### Desafio 05 ✔️

```python
# Selecionando genes e células, em quantidade de colunas iguais
df_gene = dados3.loc[:, 'g0': 'g50']
df_cel = dados3.loc[:, 'c0': 'c50']
```

```python
# Concatenando os dois DataFrames para fazer a correlação entre os genes e drogas
corr_gc = pd.concat([df_gene, df_cel], axis = 1)
```

```python
# Correlacionando genes e células
corr = corr_gc.corr().loc['g0': 'g50', 'c0': 'c50']
```

```python
# Matriz de correlação dos dados, importados do site do Seaborn. 
# https://seaborn.pydata.org/examples/many_pairwise_correlations.html
import numpy as np

sns.set_theme(style="white")
f, ax = plt.subplots(figsize=(11, 9))

# Ajustando o mapa de cores do gráfico
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
# Descrição da criação de um "heatmap"
# Matriz de correlação: https://seaborn.pydata.org/examples/many_pairwise_correlations.html

# Importação dos módulos necessários
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Definindo o tema do heatmap
sns.set_theme(style="white")

# "np.triu" retorna uma matriz triangular superior de um conjunto de dados. "np.ones_like" retorna uma matriz da ordem do conjunto 
# de dados, no caso uma matriz de "booleanos", todos True
# https://numpy.org/doc/1.20/reference/generated/numpy.triu.html?highlight=triu#numpy.triu
# https://numpy.org/doc/1.20/reference/generated/numpy.ones_like.html?highlight=ones_like#numpy.ones_like
mask = np.triu(np.ones_like(corr_gene, dtype=bool))

# Ajustando o tamanho do gráfico, como na aula passada
f, ax = plt.subplots(figsize=(15, 13))

# Configuração do gradiente de um gradiente de cores, para utilização no heatmap
# https://seaborn.pydata.org/generated/seaborn.diverging_palette.html?highlight=diverging#seaborn.diverging_palette
cmap = sns.diverging_palette(230, 20, n = 10,as_cmap=True)

# https://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap
sns.heatmap(corr_gene, mask = mask, cmap = cmap, center = 0, square = True, linewidths = 0.5, cbar_kws = {"shrink": 0.5})
plt.show()
```
